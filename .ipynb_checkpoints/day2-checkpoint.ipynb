```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(1,10,20)
y = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
plt.plot(x, y)
plt.plot(x, y2)
_:plt.plot(x, y3)

```
